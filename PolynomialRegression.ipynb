```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.linspace(-10,10,100)
Y = 0.5*X**2 + 1.5*X + np.random.randn(100)

plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

X_train = pd.DataFrame(X_train)
y_train = pd.DataFrame(y_train)
X_test = pd.DataFrame(X_test)
y_test = pd.DataFrame(y_test)
model = LinearRegression()
model.fit(X_train,y_train)

plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), color = 'r' )
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,model.predict(X_test)))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly =PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
model.fit(X_train_poly,y_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,model.predict(X_test_poly)))
```

```python
sorted_indices = X_train.iloc[:,0].values.argsort()
plt.scatter(X_train,y_train)
plt.plot(X_train.iloc[sorted_indices,0],model.predict(X_train_poly)[sorted_indices],color='r')
plt.show()
```
